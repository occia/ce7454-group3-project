```python
# Countries indexes mapping

countries = {
    "australia": "0",
    "brazil": "1",
    "canada": "2",
    "china": "3",
    "ethiopia": "4",
    "nigeria": "4",
    "germany": "5",
    "india": "6",
    "indonesia": "7",
    "iran": "8",
    "russia": "9"
}
```

```python
import os
import json
import re

country = 'iran'

country_folder = './data/processed/%s/' %country
input_path = country_folder + '/' + country + '.json'

with open(input_path, 'r') as file:
    user_profiles = json.load(file)
    for user in user_profiles:
        username = user['username']
        age = user['age']
        country_index = countries[country]
        
        image = country_folder + 'images/cut/' + username + '.jpg'
        
        if os.path.exists(image):
        # The new name is 'age_country_username'
            new_name = '%sphotos/%s_%s_%s.jpg' %(country_folder, age, country_index, username)
            print(new_name)
            os.rename(image, new_name)
```

```python
import os, os.path
import matplotlib.pyplot as plt

image_nums = {}
image_nums_all = {}

for i in range(1,101):
    if i == 0:
        pass
    if i < 10:
        i = '0' + str(i)
    data_dir = './data/processed/merged_raw/%s/' %i
    if os.path.exists(data_dir):
        os.listdir(data_dir)
        image_num = len([name for name in os.listdir(data_dir)])
        i = int(i)
        image_nums[i] = image_num
        image_nums_all[i] = image_num
    else: 
        i = int(i)
        image_nums_all[i] = 0
        
print(image_nums)

vals = image_nums

lists = sorted(image_nums_all.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()
```
